```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
toyota_csv = pd.read_csv ('Toyota_Corolla.csv')

toyo = pd.DataFrame (toyota_csv)
```

```python
toyo
```

```python
toyo.describe()
```

```python
plt.bar (toyo['NEW LISTING'], toyo['Automatic'], 10)

plt.subplot(1, 2, 1)
plt.plot(toyo['NEW LISTING'], color='purple')

plt.subplot(1, 2, 2)
plt.plot(toyo['Automatic'], color='green')

plt.show()
```
